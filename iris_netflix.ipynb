```python
import opendatasets as od
import pandas

od.download("https://www.kaggle.com/datasets/vikrishnan/iris-dataset")
od.download("https://www.kaggle.com/datasets/paramvir705/netflix-dataset")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris['species'] = iris.target
df_iris['species'] = df_iris['species'].map({
    0: 'setosa',
    1: 'versicolor',
    2: 'virginica'
})

print("Iris Dataset")
df_iris.head()
```

```python
df_netflix = pd.read_csv("/content/netflix-dataset/netflix_titles.csv")

print("Netflix Dataset")
df_netflix.head()
```

```python
print("Iris Info")
df_iris.info()

print("\nNetflix Info")
df_netflix.info()
```

```python
df_iris.hist(figsize=(10,8), bins=20)
plt.suptitle("Iris Dataset – Numerical Feature Distribution")
plt.show()
```

```python
df_netflix[['release_year']].hist(bins=30, figsize=(6,4))
plt.title("Netflix Dataset – Release Year Distribution")
plt.show()
```

```python
sns.countplot(x='species', data=df_iris)
plt.title("Iris Species Distribution")
plt.show()
```

```python
sns.countplot(x='type', data=df_netflix)
plt.title("Netflix Content Type Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df_iris.drop('species', axis=1))
plt.title("Iris Dataset – Outlier Detection")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df_iris.drop('species', axis=1))
plt.title("Iris Dataset – Outlier Detection")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df_iris.drop('species', axis=1).corr(),
            annot=True, cmap='coolwarm')
plt.title("Iris Dataset – Correlation Heatmap")
plt.show()
```

```python
df_netflix['type_encoded'] = df_netflix['type'].map({'Movie': 0, 'TV Show': 1})

plt.figure(figsize=(5,4))
sns.heatmap(df_netflix[['release_year', 'type_encoded']].corr(),
            annot=True, cmap='Blues')
plt.title("Netflix Dataset – Correlation Heatmap")
plt.show()
```
